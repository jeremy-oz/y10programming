# Analyse time and data

We are usually interested in everything in the data set.

Can I just see the prices of year 2016?

Steps:
1. Get the year from the index: `stocks.index.year`
1. Produce a mask: `mask2016 = (stocks.index.year == 2016)`
1. Apply the filter to stocks and save the result to stocks2017: `stocks2016 = stocks[mask2016]`

The mask is a special trick provided by pandas.


```python
mask2016 = (stocks.index.year == 2016)
stocks2016 = stocks[mask2016]
stocks2016 # if we just put a vairable here, its content will be displayed.
```

Let see the opening and closing prices for 2016.

It actually only has 195 days of data.

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(stocks2016.Open)
ax.plot(stocks2016.Close)
;
```

# Exercise 2: Produce a chart showing highest and lowest prices for 2015.

```python
mask2015 = (stocks.index.year == 2015)
stocks2015 = stocks[mask2015]
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(stocks2015.Open)
ax.plot(stocks2015.Close)
;
```

# Finally, we can create charts providing multiple information for investors.

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(8, 5), sharex=True)

ax1.plot("High", data=stocks, color='green')
ax1.plot("Low", data=stocks, color='red')
ax2.plot("Volume", data=stocks)
ax2.fill_between(stocks.index, 0, "Volume", data=stocks, alpha=.25)
ax2.set_ylim(0)
ax1.legend()
ax2.legend()

plt.tight_layout()
```
