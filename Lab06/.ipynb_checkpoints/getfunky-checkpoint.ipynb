## Data Structure

* list
    Example: numbers = [11, 23, 15, 17, 18]
* dict
    Example: scores = {'Liam': 13, 'Alex': 18, 'Daniel': 20, 'Josh': 22}

## Add members into a list and dict

### list

fruits = ['apple', 'banana', 'orange']
* Add a member using append(): fruits.append('mango')
* Remove a member using remove(): fruits.remove('banana')
* Remove the last member using pop(): fruits.pop()

### dict

diet = {}
* Create a new dict: diet = {'fruit': 'apple', 'staple': 'pasta', 'snake': 'yoghourt', 'meals':3}
* Add a member: diet['drink'] = 'coke'
* Update a member: diet['drink'] = 'water'
* Remove a member using: my_dict.pop('drink')

### Example: Guess a number


````python
scores = {}

scores['John'] = 89
scores['Tim'] = 99
scores['Jimmy'] = 78

print(scores)

for student, score in scores.items():
    print(student, 'gets', score)
````

# Get a number: funky version

You should already know this program well.

Now we are going to make a funky version by adding ascii art.

```

import random

print('---------------------------------')
print('   GUESS THAT NUMBER GAME')
print('---------------------------------')
print()

the_number = random.randint(0, 100)
guess = -1

name = input('Player what is your name? ')

while guess != the_number:
    guess_text = input('Guess a number between 0 and 100: ')
    guess = int(guess_text)

    if guess < the_number:
        # print('Your guess of ' + guess + ' was too LOW.')
        print('Sorry {}, your guess of {} was too LOW.'.format(name, guess))
    elif guess > the_number:
        print('Sorry {}, your guess of {} was too HIGH.'.format(name, guess))
    else:
        print('Excellent work {}, you won, it was {}!'.format(name, guess))

print('done')
```
